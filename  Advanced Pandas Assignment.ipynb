```python
# Monthly saled trends
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

salesandcustomer_file_path = 'C:\\Users\\kaberac\\Desktop\\Data Projects &Assignment'
customer_df = pd.read_csv(salesandcustomer_file_path + '\\customer_data.csv')
sales_df = pd.read_csv(salesandcustomer_file_path + '\\sales_data.csv')

merged_df = pd.merge(sales_df, customer_df, on='customer_id', how='left')
merged_df['sales_date'] = pd.to_datetime(merged_df['sales_date'], errors='coerce')

# Drop rows with missing values
merged_df.dropna(inplace=True)

# Ensure 'unit_price' is numeric
merged_df['unit_price'] = pd.to_numeric(merged_df['unit_price'], errors='coerce')

# Check for any non-numeric values in 'unit_price'
non_numeric_prices = merged_df[merged_df['unit_price'].isna()]
#print("Non-numeric 'unit_price' values:\n", non_numeric_prices)

# Remove rows with non-numeric 'unit_price'
merged_df.dropna(subset=['unit_price'], inplace=True)
# Extract the month from 'sales_date'
merged_df['month'] = merged_df['sales_date'].dt.to_period('M')
# Group by month and calculate total sales amount
monthly_sales = merged_df.groupby('month')['unit_price'].sum().reset_index()

# Ensure 'month' is converted back to string for plotting purposes
monthly_sales['month'] = monthly_sales['month'].astype(str)

# Plot the monthly sales trends
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='month', y='unit_price')
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Total Unit Price')
plt.xticks(rotation=45)
plt.show()
```

```python
#Analyze customer retention rates over time.
import pandas as pd

salesandcustomer_file_path = 'C:\\Users\\kaberac\\Desktop\\Data Projects &Assignment'
sales_df = pd.read_csv(salesandcustomer_file_path + '\\sales_data.csv')

# Convert sales_date to datetime format
sales_df['sales_date'] = pd.to_datetime(sales_df['sales_date'])

# Extract year and month from sales_date
sales_df['year_month'] = sales_df['sales_date'].dt.to_period('M')

# Determine the first purchase date for each customer
first_purchase = sales_df.groupby('customer_id')['sales_date'].min().reset_index()
first_purchase.columns = ['customer_id', 'first_purchase_date']
first_purchase['first_purchase_period'] = first_purchase['first_purchase_date'].dt.to_period('M')

# Merge first purchase date with sales data
sales_df = sales_df.merge(first_purchase, on='customer_id')

# Create a DataFrame to track retention
retention_df = sales_df[['customer_id', 'year_month', 'first_purchase_period']].drop_duplicates()

# Calculate retention by period
retention_df['cohort_period'] = (retention_df['year_month'] - retention_df['first_purchase_period']).apply(lambda x: x.n)
retention_counts = retention_df.groupby(['first_purchase_period', 'cohort_period']).size().unstack(fill_value=0)

# Calculate retention rates
retention_rates = retention_counts.div(retention_counts.iloc[:, 0], axis=0)

# Display the retention rates
print(retention_rates)

# Plot the retention rates
plt.figure(figsize=(12, 8))
sns.heatmap(retention_rates, annot=True, fmt='.0%', cmap='Blues')
plt.title('Customer Retention Rates Over Time')
plt.xlabel('Cohort Period (Months)')
plt.ylabel('First Purchase Period')
plt.show()

```

```python
#Implement time series analysis techniques (e.g., moving averages, exponential smoothing) to forecast future sales trends.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.holtwinters import ExponentialSmoothing

salesandcustomer_file_path = 'C:\\Users\\kaberac\\Desktop\\Data Projects &Assignment'
sales_df = pd.read_csv(salesandcustomer_file_path + '\\sales_data.csv')

# Convert sales_date to datetime format
sales_df['sales_date'] = pd.to_datetime(sales_df['sales_date'])

# Extract year and month from sales_date
sales_df['year_month'] = sales_df['sales_date'].dt.to_period('M')

# Aggregate sales data by month and year
monthly_sales = sales_df.groupby('year_month')['sales_quantity'].sum().reset_index()
monthly_sales.set_index('year_month', inplace=True)
monthly_sales.index = monthly_sales.index.to_timestamp()

# Calculate Simple Moving Average (SMA)
monthly_sales['SMA_3'] = monthly_sales['sales_quantity'].rolling(window=3).mean()
monthly_sales['SMA_6'] = monthly_sales['sales_quantity'].rolling(window=6).mean()

# Calculate Exponential Moving Average (EMA)
monthly_sales['EMA_3'] = monthly_sales['sales_quantity'].ewm(span=3, adjust=False).mean()
monthly_sales['EMA_6'] = monthly_sales['sales_quantity'].ewm(span=6, adjust=False).mean()

# Plot Moving Averages
plt.figure(figsize=(14, 7))
plt.plot(monthly_sales['sales_quantity'], label='Monthly Sales')
plt.plot(monthly_sales['SMA_3'], label='3-Month SMA')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

salesandcustomer_file_path = 'C:\\Users\\kaberac\\Desktop\\Data Projects &Assignment'
sales_df = pd.read_csv(salesandcustomer_file_path + '\\sales_data.csv')

# Convert sales_date to datetime format
sales_df['sales_date'] = pd.to_datetime(sales_df['sales_date'])

# Add a mock promo column
# For simplicity, let's randomly assign 50% of the sales as promotional
np.random.seed(42)
sales_df['promo'] = np.random.choice([True, False], size=len(sales_df))

# Separate the data into control and test groups
control_group = sales_df[sales_df['promo'] == False]
test_group = sales_df[sales_df['promo'] == True]

# Calculate sales metrics
control_sales = control_group['sales_quantity'].sum()
test_sales = test_group['sales_quantity'].sum()
control_mean = control_group['sales_quantity'].mean()
test_mean = test_group['sales_quantity'].mean()

# Perform statistical testing
t_stat, p_value = stats.ttest_ind(control_group['sales_quantity'], test_group['sales_quantity'], equal_var=False)

print("Control Group Total Sales:", control_sales)
print("Test Group Total Sales:", test_sales)
print("Control Group Average Sales:", control_mean)
print("Test Group Average Sales:", test_mean)
print("T-statistic:", t_stat)
print("P-value:", p_value)

# Visualize the results
plt.figure(figsize=(12, 6))

# Plot total sales comparison
plt.subplot(1, 2, 1)
sns.barplot(x=['Control', 'Test'], y=[control_sales, test_sales])
plt.title('Total Sales Comparison')
plt.ylabel('Total Sales Quantity')

# Plot average sales comparison
plt.subplot(1, 2, 2)
sns.barplot(x=['Control', 'Test'], y=[control_mean, test_mean])
plt.title('Average Sales Comparison')
plt.ylabel('Average Sales Quantity')

plt.tight_layout()
plt.show()

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("The difference between control and test groups is statistically significant (p < 0.05).")
else:
    print("The difference between control and test groups is not statistically significant (p >= 0.05).")
```
